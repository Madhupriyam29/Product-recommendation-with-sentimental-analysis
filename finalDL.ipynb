```python
import pandas as pd
from textblob import TextBlob

# Load the data
df = pd.read_csv('C:/Users/Madhu/Downloads/lapcleaned_detail.csv')

# Function to get sentiment
def get_sentiment(performance_review):
    if isinstance(performance_review, str):  # Check if the review is a string
        analysis = TextBlob(performance_review)
        if analysis.sentiment.polarity > 0:
            return 'Positive'
        elif analysis.sentiment.polarity < 0:
            return 'Negative'
        else:
            return 'Neutral'
    else:
        return 'Neutral'  # Handle non-string entries

# Apply the function to the review column
df['Sentiment'] = df['performance_review'].apply(get_sentiment)

# Save the updated DataFrame to a new CSV file
df.to_csv('C:/Users/Madhu/Downloads/lapsentiment_detail.csv', index=False)

# Display the first few rows to verify
print(df[['performance_review', 'Sentiment']].head())
```

```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('C:/Users/Madhu/Downloads/lapsentiment_detail.csv')

# Define the sentiment score mapping
sentiment_scores = {
    'Positive': 1,
    'Neutral': 0,
    'Negative': -1
}

# Map the Sentiment column to sentiment scores
df['Sentiment_Score'] = df['Sentiment'].map(sentiment_scores)

# Display the first few rows to verify
print(df.head())

# Optionally, save the updated DataFrame to a new CSV
df.to_csv('C:/Users/Madhu/Downloads/lap_reviews_with_scores.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'Rating' is a column in the DataFrame
plt.figure(figsize=(8, 6))
sns.countplot(x='performance_review', data=df, palette='viridis')
plt.title('Distribution of performance_review')
plt.xlabel('performance_review')
plt.ylabel('Count')
plt.show()
```
