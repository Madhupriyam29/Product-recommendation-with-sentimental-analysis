```python
#oppo phone review

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# Initialize the WebDriver with error handling
def initialize_driver():
    try:
        driver = webdriver.Chrome()
        driver.maximize_window()
        return driver
    except Exception as e:
        print(f"Failed to initialize WebDriver: {e}")
        return None

driver = initialize_driver()
wait = WebDriverWait(driver, 10)

# Initialize an empty list to store review details
review_details = []

def wait_for_overlay_to_disappear():
    try:
        wait.until(EC.invisibility_of_element((By.CLASS_NAME, 'H30l+S')))
    except Exception:
        pass

try:
    driver.get('https://www.flipkart.com/oppo-k12x-5g-45w-supervooc-charger-in-the-box-breeze-blue-128-gb/product-reviews/itma30cb38861d4c?pid=MOBH2Q4PN6FWZCGV&lid=LSTMOBH2Q4PN6FWZCGVPMAT2I&sortOrder=MOST_HELPFUL&certifiedBuyer=false&aid=overall')
    
    wait.until(EC.presence_of_element_located((By.TAG_NAME, 'body')))

    page_number = 1

    while True:
        try:
            # Wait for pagination container to load
            pagination_container = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="container"]/div/div[3]/div/div/div[2]/div[13]/div/div')))

            next_page_button = wait.until(EC.element_to_be_clickable((By.XPATH, f'.//a[contains(@class,"cn++Ap") and text()="{page_number + 1}"]')))
            driver.execute_script("arguments[0].scrollIntoView(true);", next_page_button)
            time.sleep(1)
            wait_for_overlay_to_disappear()

            # Click the next page button
            for _ in range(3):
                try:
                    driver.execute_script("arguments[0].click();", next_page_button)
                    print(f'Navigated to page {page_number + 1}')
                    time.sleep(5)
                    break
                except Exception as e:
                    print(f"Click failed: {e}")
                    time.sleep(1)
            else:
                print(f"Failed to click the button after retries.")
                break

            # Scroll down to load all reviews
            old_page_height = driver.execute_script("return document.body.scrollHeight")

            product_element_name = wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class="Vu3-9u eCtPz5"]')))
            product_name = product_element_name.text.strip()
            print(f"Extracted Product Name: {product_name}")

            while True:
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(2)
                new_page_height = driver.execute_script("return document.body.scrollHeight")
                if new_page_height == old_page_height:
                    break
                old_page_height = new_page_height

            # Wait for review items to load
            review_container = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//div[@class="col EPCmJX Ma1fCG"]')))

            for review in review_container:
                try:
                    overall_review = review.find_element(By.XPATH, './/div[@class="ZmyHeo"]').text
                    rating = review.find_element(By.CLASS_NAME, "XQDdHH.Ga3i8K").text
                    review_details.append({
                        'product_name': product_name, 
                        'review': overall_review,
                        'rating': rating
                    })

                except Exception as e:
                    print(f"Failed to extract details from review: {e}")
                    continue

            page_number += 1

        except Exception as e:
            print(f"Failed to navigate to page {page_number + 1}: {e}")
            break

    # Create a DataFrame from review_details list
    oppo_reviews = pd.DataFrame(review_details)
    print(oppo_reviews)

except Exception as e:
    print(f"Error occurred: {e}")

finally:
    if driver:
        driver.quit()
```
